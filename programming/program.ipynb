```python
#사용자가 입력한 게임 회수만큼 로또 번호를 출력하는 프로그램을 작성하세요(1~45) 6개 숫자
import random
innum = int(input('플레이 할 게임 횟수를 입력해 주세요 : ')) - 1
while 1:
    a = int(random.randrange(1,46))
    b = int(random.randrange(1,46))
    c = int(random.randrange(1,46))
    d = int(random.randrange(1,46))
    e = int(random.randrange(1,46))
    f = int(random.randrange(1,46))
    print('{}번 {}번 {}번 {}번 {}번 {}번'.format(a,b,c,d,e,f))
    if innum == 0:
        break
    else:
        innum -= 1
```

```python
#남녀 파트너 정해주기 프로그램을 작성하세요(5개 커플)
import random
hm = int(input('몇 개의 커플을 생성하시겠습니까? : ')) * 2

male = []
female = [] 
while True:
    if hm == 0:
        break
    else:
        gender = input('성별을 입력해 주세요 : ')
        
        if gender == '남자':
            name = input('이름을 입력해 주세요 : ')
            male.append(name)
            hm -= 1
        elif gender == '여자':
            name = input('이름을 입력해 주세요 : ')
            female.append(name)
            hm -= 1
        else :
            print('성별을 올바르게 입력해 주세요')
random.shuffle(male)
random.shuffle(female)
print(male[0],female[0],'가 첫 번째 커플입니다')
print(male[1],female[1],'가 두 번째 커플입니다')
print(male[2],female[2],'가 세 번째 커플입니다')
print(male[3],female[3],'가 네 번째 커플입니다')
print(male[4],female[4],'가 다섯 번째 커플입니다')
```

```python
# Q13. 음식점의 음식 주문가격 산출 프로그램을 자유롭게 작성하세요.
class FastFood:
    bill = 0
    
    def __init__(self,a):
        print("====================")
        print("1 : 김밥     | 2000원")
        print("2 : 짜장라면 | 4000원")
        print("3 : 봉지라면 | 5000원")
        print("4 : 우동     | 6000원")
        print("5 : 라볶이   | 6000원")
        print("====================")
        a = input("주문하실 메뉴를 입력해주세요 : ")
    
    def start(self,a):
        if a == 1 or a == 김밥:
            print("김밥을 선택했습니다")
            ea1 = int(input('몇 개를 주문하시겠습니까? : '))
            bill = (2000 * ea1)
            print("결제 금액은 {}원 입니다".format(bill))
        elif a == 2 or a == 짜장라면:
            print("짜장라면을 선택했습니다")
            ea2 = int(input('몇 개를 주문하시겠습니까? : '))
            bill = (4000 * ea2)
            print("결제 금액은 {}원 입니다".format(bill))
        elif a == 3 or a == 봉지라면:
            print("봉지라면을 선택했습니다")
            ea3 = int(input('몇 개를 주문하시겠습니까? : '))
            bill = (5000 * ea3)
            print("결제 금액은 {}원 입니다".format(bill))
        elif a == 4 or a == 우동:
            print("우동을 선택했습니다")
            ea4 = int(input('몇 개를 주문하시겠습니까? : '))
            bill = (6000 * ea4)
            print("결제 금액은 {}원 입니다".format(bill))
        elif a == 5 or a == 라볶이:
            print("라볶이를 선택했습니다")
            ea5 = int(input('몇 개를 주문하시겠습니까? : '))
            bill = (4000 * ea5)
            print("결제 금액은 {}원 입니다".format(bill))
        else:
            print('없는 메뉴 입니다!')
    
print("====================")
print("1 : 김밥     | 2000원")
print("2 : 짜장라면 | 4000원")
print("3 : 봉지라면 | 5000원")
print("4 : 우동     | 6000원")
print("5 : 라볶이   | 6000원")
print("====================")
a = input("주문하실 메뉴를 입력해주세요 : ")
    
start = FastFood(1)
start.start(a)
```
