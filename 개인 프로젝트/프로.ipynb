```python
import datetime
import getpass
import re
import os
from bs4 import BeautifulSoup
import urllib


class Start:
    def __init__(self):
        print("프로그램이 실행 되었습니다.")
        print("*"*50)
    def login(self):
        while True:
            print("*"*50)
            p = input("1. 로그인, 2. 회원가입, 3. 종료 : ")
            print("*"*50)
            p_data = {}
            umm = []
            if p == '1' or p == '로그인' : # 로그인 프로세스 시작
                p_id = str(input("사용자 ID : "))
                p_pw = getpass.getpass("사용자 PW : ")
                with open('logdata.txt','a') as f: #로그데이터에 행위 기록
                    f.write("logdata = {}  로그인 시도 ID = {}\n".format(datetime.datetime.now(),p_id))
                with open('memberdata.txt','r') as f: #회원데이터에 ID 있는지 확인
                    data = f.read() 
                    id_filter = re.compile('ID\s:\s%s+'%p_id) #id 가 데이터 파일에 있는지 확인
                    id_f_data = id_filter.findall(data)
                    id_f2_data = ':'.join(id_f_data)
                    id_data = id_f2_data.replace('ID : ','')
        
                    
                    id_pw_filter = re.compile('ID\s:\s%s\s/\sPW\s:\s\w+'%id_data) # 데이터 파일에 id 기반으로 pw 가 있는지 확인
                    pw_f_data = id_pw_filter.findall(data)
                    pw_f2_data = '/'.join(pw_f_data)
                    pw_data = pw_f2_data.replace('ID : %s / PW : '%id_data,'')
            
                    if p_id == id_data and p_pw == pw_data:
                        with open('logdata.txt','a') as f: #로그데이터에 행위 기록
                            f.write("logdata = {}  로그인 완료 ID = {}\n".format(datetime.datetime.now(),p_id))
                        path = 'datafile'
                        file_list = os.listdir(path)
                        for i in file_list:
                            i = i.replace('.txt','')
                               
                                
                        print("""
                        ┌──────────────────────────────────┐
                        │   오늘의 뉴스 or 주식
                        ├──────────────────────────────────┤
                        │   코로나 확진자 현황 : API 시리얼 키 승인 대기 중
                        ├──────────────────────────────────┤
                        │   {}님 환영합니다  
                        ├──────────────────────────────────┤
                        │   게시물 작성 = New/ 삭제 = Del (enter->) 게시물 제목              │ 
                        │   게시물 리스트 = Golist                                           │
                        │   게시물 보기 = Move (enter->) 게시물 번호                         │
                        │   게시물 추천 = Recommand (enter->) 추천 or 비추천 입력            │
                        │   로그아웃 = Logout                                                │
                        │   회원탈퇴 = Delid                                                 │
                        ├──────────────────────────────────┤
                        │                              인기 게시물                           │
                        ├──────────────────────────────────┤
                        │   제목 : {}
                        └──────────────────────────────────┘
                        
                        """.format(p_id,i))
                        while True:
                            wow_a = input("명령어를 입력하세요. : ")
                            print("*"*50)
                
                            if wow_a == 'New' or wow_a == 'new' or wow_a == '작성':
                                print("         게시물을 작성합니다")
                                print("┌──────────────────────────────────┐")
                                title = input("""     게시물의 제목을 입력해주세요 : """)
                                print("├──────────────────────────────────┤")
                                text = input("""     게시물의 내용을 입력해주세요 : """)
                                print("└──────────────────────────────────┘")
                                filename = 'datafile/%s.txt' %title
                                with open('logdata.txt','a') as f:
                                    f.write("logdata = {}  게시물 작성 ID = {} 게시물 제목 = {}\n".format(datetime.datetime.now(),p_id,title))
                                with open(filename,'a')as f:
                                    f.write('{}'.format(text))
                            elif wow_a == 'DelNumber' or wow_a == 'delnumber'or wow_a == '삭제':
                                print("게시물을 삭제합니다")
                                rm_title = str(input('삭제 할 게시물 제목을 입력하세요'))
                                choice = str(input("삭제하시려면 삭제합니다를 입력해주세요 : "))
                                if choice == '삭제합니다':
                                    os.remove(r'datafile/%s.txt' %rm_title)
                                    print('게시물이 삭제 되었습니다.')
                                else :
                                    pass
                            elif wow_a == 'Move'or wow_a == 'move'or wow_a == '이동':
                                pr_title = str(input(('이동할 게시물의 제목을 입력해 주세요 : ')))
                                print("게시물로 이동합니다")
                                with open('datafile/{}.txt'.format(pr_title)) as pr_data:
                                    rd_data = pr_data.read()
                                    print("""
                                    ┌──────────────────────────────────┐
                                    │   게시물 제목 : {}
                                    ├──────────────────────────────────┤
                                    │   게시물 내용 : {}
                                    └──────────────────────────────────┘
                                    """.format(pr_title, rd_data))
                            elif wow_a == 'Golist'or wow_a == 'golist'or wow_a == '리스트':
                                print("게시물 리스트를 출력합니다")
                                path = 'datafile'
                                file_list = os.listdir(path)
                                for i in file_list:
                                    i = i.replace('.txt','')
                                    print("""
                                    ├──────────────────────────────────┤
                                    │    게시물 제목 : %s""" %i)
                                
                            elif wow_a == 'Recommand'or wow_a == 'recommand'or wow_a == '평가':
                                print("게시물을 평가합니다")
                            elif wow_a == 'Logout'or wow_a == 'logout'or wow_a == '로그아웃':
                                print("로그아웃 합니다")
                                break
                            elif wow_a == 'Delid' or wow_a == 'delid' or wow_a == '회원탈퇴':
                                print("회원탈퇴를 진행합니다")
                                real = input("정말 회원탈퇴를 진행하시겠습니까? : ")
                                if real == 'yes' or real == '네':
                                    print("회원탈퇴가 완료되었습니다")
                                else:
                                    print("초기 화면으로 돌아갑니다")
                            else:
                                print("명령어가 올바르지 않습니다")
                    else:
                        print("아이디 패스워드를 확인해주세요")
                
            elif p == '2' or p == '회원가입' : # 회원 가입 프로세스 시작
                while True:
                    with open('memberdata.txt','r') as f: #회원데이터에 있는지 확인
                        data = f.read()
                    id_filter = re.compile('ID\s:\s\w+')
                    id_f_data = id_filter.findall(data)
                    id_f2_data = ':'.join(id_f_data)
                    id_f3_data = id_f2_data.replace('ID : ','')
                    p_data = id_f3_data.split(':')
                    p_id = input("가입 할 ID : ")
                    if p_id in p_data: # 중복된 아이디는 생성이 안되도록 조건
                        print("이미 사용하고 있는 ID 입니다.")
                        with open('logdata.txt','a') as f:
                            f.write("logdata = {}  회원가입 아이디 중복확인 실패 ID = {}\n".format(datetime.datetime.now(),p_id))
                    else:
                        print("사용 할 수 있는 ID 입니다")
                        p_pw = input("사용 할 PW : ")
                        with open('memberdata.txt','a') as f:
                            f.write("{} / ID : {} / PW : {}\n".format(datetime.datetime.now(),p_id, p_pw))
                        with open('logdata.txt','a') as f:
                            f.write("logdata = {}  회원가입 아이디 중복확인 성공 ID = {}\n".format(datetime.datetime.now(),p_id))
                        print("회원가입이 완료 되었습니다")
                        break
                        with open('logdata.txt','a') as f:
                            f.write("logdata = {}  회원가입 완료 ID = {}\n".format(datetime.datetime.now(),p_id))
                    break
                       
    
            elif p == '3' or p == '종료': # 종료
                print("프로그램을 종료합니다")
                print("*"*50)
                break
            elif p == '바보' or p == '멍청이':    
                print("기분이 안좋아져서 종료합니다.")
                print("*"*50)
                break
            else:
                print("메뉴 중 하나를 골라주세요")
```
